### Preproceso

Primero voy a indetificar la muestra y el objetivo:

```python
y = df["Class"]  # Atención, hay que asignar y primero si utilizo el drop en vez de el iloc porque no se puede asignar una columna ya dropped.
X = df.drop(["Class"], axis = 1)
```

```python
y
```

```python
X
```

Ahora puedo dividir la muestra y el objetivo para crear el entorno de entrenamiento y el de test para mis modelos de regresion:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

Finalmente voy a normalizar la muestra pues tengo valores de cientos (p.e. la cantidad de Magnesio) y valores de decimales (p.e. la cantidad de Nonflavonoids_phenols):

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

Ahora ya puedo trabajar con los distintos modelos de clasificacion:

### 1.1. Decision tree 

```python
from sklearn.tree import DecisionTreeClassifier # Importar Decision Tree Classifier
from sklearn import metrics #Importar scikit-learn metrics module para el cálculo de acccuracy
```

```python
# Creo el Decision tree clasifier objeto
clf = DecisionTreeClassifier()
# TLo entreno
clf = clf.fit(X_train_scaled,y_train)
#Lanzo la predición para el data test
y_pred = clf.predict(X_test_scaled)
```

```python
# Evaluo la accuracy del modelo, es decir qué porcentaje de veces es el modelo correcto
# Evaluo la precision del modelo, es decir, qué porcentaje de veces cuando dice que es de una clase es así.
accuracy_tree = metrics.accuracy_score(y_test, y_pred)
precision_tree = metrics.precision_score(y_test, y_pred, average='weighted')
print("Accuracy:", accuracy_tree)
print("Precision:", precision_tree)
print("\033[1mTengo una Accuracy del:", round(accuracy_tree*100, 2), "%\033[0m")
print("\033[1mTengo una Precision del:", round(precision_tree*100, 2), "%\033[0m")
```

```python
#Y la profundidad del arbol
depth = clf.get_depth()
print(f"Profundidad del árbol: {depth}")
```

Visualicemos el modelo:

```python
from sklearn.tree import plot_tree
feature_cols = df.columns.drop(["Class"])
plt.figure(figsize=(15, 15))
plot_tree(clf, feature_names=feature_cols, class_names="Class", filled=True, rounded=True)
plt.show()
```

**Calcularé la matriz de confusión**:

```python
from sklearn.metrics import confusion_matrix
# Calculo la matriz de confusión
confusion_tree = confusion_matrix(y_test, y_pred)

# Imprimo la matriz de confusión
print("Matriz de confusión:")
print(confusion_tree)
```

La dibujo en un heatmap para verla mas claramente:

Tengo una accuracy y una precision razonables, pero emjorables, y se observa en la matriz. veamos como puedo mejorar el modelo.

**Voy a somerter el modelo a un proceso de pruning con el objetivo de identificar la profundidad optima, que consigue reducir la varianza y aumentar la capacidad predictiva del modelo**:

```python
# Pruning por validación cruzada

from sklearn.model_selection import GridSearchCV

# Valores de ccp_alpha evaluados
param_grid = {'ccp_alpha':np.linspace(0, 5, 10)}

# Búsqueda por validación cruzada
grid = GridSearchCV(
        # El árbol se crece al máximo posible antes de aplicar el pruning
        estimator = DecisionTreeClassifier(
                            max_depth         = None,
                            min_samples_split = 2,
                            min_samples_leaf  = 1,
                            random_state      = 1
                       ),
        param_grid = param_grid,
        scoring    = 'accuracy',
        cv         = 10,
        refit      = True,
        return_train_score = True
      )

grid.fit(X_train_scaled, y_train)

fig, ax = plt.subplots(figsize=(6, 3.84))
scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_ccp_alpha', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_ccp_alpha', y='mean_test_score', yerr='std_test_score', ax=ax)
ax.set_title("Error de validacion cruzada vs hiperparámetro ccp_alpha");
```

de donde como se pude ver:

```python
grid.best_params_
```

Indetificado el valor mejor de ccp_alpha se debería reentrenar el Tree Clasifier con este valor, pero como he puesto en el GridSearchCV() refit=True, el modelo se ha reentrenado automáticamente y el modelo tiene los parámetros alamcenados en .best_estimator_, así que creo el nuevo modelo con estos parámetros:

```python
modelo_final = grid.best_estimator_
print(f"Profundidad del árbol: {modelo_final.get_depth()}")
print(f"Número de nodos terminales: {modelo_final.get_n_leaves()}")
```

```python
y_pred1 = modelo_final.predict(X_test_scaled)
```

```python
# Evaluo la accuracy del modelo, es decir qué porcentaje de veces es el modelo correcto
# Evaluo la precision del modelo, es decir, qué porcentaje de veces cuando dice que es de una clase es así.
accuracy_tree1 = metrics.accuracy_score(y_test, y_pred1)
precision_tree1 = metrics.precision_score(y_test, y_pred1, average='weighted')
print("Accuracy:", accuracy_tree1)
print("Precision:", precision_tree1)
print("\033[1mTengo una Accuracy del:", round(accuracy_tree1*100, 2), "%\033[0m")
print("\033[1mTengo una Precision del:", round(precision_tree1*100, 2), "%\033[0m")
```

**Voy a encontrar el mejor K para la clasificación usando el f1-score**:

```python
from sklearn.metrics import f1_score

f1s = []

# Calcula el f1 score para K valores entre 1 y 40
for i in range(1, 8):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled, y_train)
    pred_i = knn.predict(X_test_scaled)
    # usando average='weighted' para calcualr la media podnerada de las 3 clases 
    f1s.append(f1_score(y_test, pred_i, average='weighted'))
```

Grafico el resultado:

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 8), f1s, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('F1 Score K Value')
plt.xlabel('K Value')
plt.ylabel('F1 Score')
```

Podemos ver que precismente el 5 da el valor mas bajo de K, asi que lo recalculo con valor de 6:

## Compara els models de classificació utilitzant la precisió (accuracy), una matriu de confusió i d’altres mètriques més avançades.

La clasificación por precisión no siempre nos asegura que el modelo sea suficientemente bueno porque depende también de la distribución de los valores que hay debajo (es decir, llevado al absurdo, si tuvieramos p.e. que 99 y_test resutls dan 1 y uno da 0), un modelo que dierasiempre 1, nos daría un 99% de accuracy y pensaríamos que es bueno, porque no estamos viendo los falsos positivos, falsos negatiovs, y verdaderos negativos del modelo. 

Eso nos lo da la matriz de Confusión que nos permite ver ademas de el error de accuracy (Precisión) los siguientes errores:

**Accuracy/Exactitud general del modelo**: En general, en qué % de las veces la predicción es correcta?

$$Precisión general del modelo/Accuracy = \large\tfrac{(TP + TN)}{(TP + TN + FP + FN)}$$  
<center>print(metrics.accuracy_score(y_test, y_pred_class))  
    
 _  
**Classification Error/Miclassification rate**: EN general, en qué % de las veces la predicción es incorrecta?
    
$$CassificationError = \large\tfrac{(FP + FN)}{(TP + TN + FP + FN)}$$
<center>print(1 - metrics.accuracy_score(y_test, y_pred_class))  
    
_  
**Sensibilidad/Tasa de verdaderos positivos/Recall**: Cunado el valor real es positivo (clase = 1, p.e.), en qué % de las veces la predicción es correcta?, (cuantos verdaderos positivos del total de positivos), Cuan sensible es prediciendo positivos?.  
    
$$Sensibilidad = \large\tfrac{(TP)}{(TP + FN)}$$  
<center>print(metrics.recall_score(y_test, y_pred_class))  
    
_   
**Specificity/Selectivity**: Cuando el valor real es negativo (clase = 0, p.e.), en qué % de las veces la predicción es correcta?, (cuantos verdaderos negativos del total de negativos), Cuán específico o selectivo es prediciendo negativos?)  
    
$$Specificity = \large\tfrac{(TN)}{(TN + FP)}$$
<center>print(specificity)  
    
_  
**False Positive Rate**: Cuando el valore real es negativo, que % de las veces la predicción es incorrecta?  
    
$$FalsePositiveRate = \large\tfrac{(FP)}{(FP + TN)}$$  
<center>print(1 - specificity)  
    
_   
**Precision**: Calcula la calidad de las predicciones positivas. Cuando predice un positivo, que % de las veces la predicción es correcta? Cuán preciso es cuando predice valores positivos?  (En caso de problems multicalse se puede calcular el valor de cada clase y calcualr el promedio (miccro, macro, o weighted), donde:  
- average='micro': Calcula la precisión globalmente contando el total de verdaderos positivos, falsos positivos y falsos negativos.  
- average='macro': Calcula la precisión para cada clase y luego calcula el promedio sin tener en cuenta el desequilibrio de clases.  
- average='weighted': Calcula la precisión para cada clase y luego calcula el promedio ponderado por el número de muestras en cada clase.  
    
$$Precision = \large\tfrac{(TP)}{(TP + FP)}$$  
<center>print(metrics.precision_score(y_test, y_pred_class))  
    
_  
Además de estas hay muchas otras metricas que se pueden evaluar: F1 score, Matthews correlation cefficiente, etc.    
   
**F1-Score**: Es una medida que combina la precisión y el recall para proporcionar una medida general del rendimiento del modelo. Se calcula como la media armónica de la precisión y el recall. Es útil cuando se desea tener un equilibrio entre la precisión y la capacidad de recuperar correctamente las muestras positivas.  
    
**Cuales verificar?**:    
    Elegir las metricas en función del objetivo  del análisis:  
    - Cual es mas imporntante reducir? FP o FN  
    - Elegir metricas con la variable relevante (FP o FN)  
    
**Casos**:  
    - Filtrado de Spam (Siendo la clase positiva = spam):  
    Optimizar para precisión o Specificity (en ambos FP son variables), porque nos interesa mas que haya falsos negativos (que vayan al inbox y detectaremos) que que haya falsos positivos (que vayan a la carpeta spam y los perdamos).  
    - Operaciones fraudulentas (siendo la clase positiva = fraudulenta):  
    Optimizar para Sensitivity (donde FN es una variable), porque FP (transacciones correctas identificadas como fraudulentas) serán estudiadas y validadas, pero FN ( operaciones fraudulentas dadas por legales) las dejaríamos pasar.
